```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

info = pd.read_csv('Internet-Demographic.csv')
info
```

<h1 style='color:blue'>Insight 1</h1>
<h3>The histogram of birth rates shows how birth rates are distributed among countries. If the distribution is skewed right, most countries have lower birth rates, whereas a normal or uniform distribution suggests more balanced variations across nations.</h3>

```python
plt.figure(figsize=(8, 5))
sns.histplot(info['Birth rate'], bins=15, kde=True, color='blue')
plt.title('Distribution of Birth Rates')
plt.xlabel('Birth Rate')
plt.ylabel('Frequency')
plt.show()
```

<h1 style='color:blue'>Insight 2</h1>
<h3>The top 5 countries with the highest internet usage‚ÄîIceland, Bermuda, Norway, Sweden, and Denmark‚Äîindicate regions with strong digital infrastructure. Their high internet penetration suggests advanced economies and widespread access to online services.</h3>

```python
top_5_countries = info.nlargest(5, 'Internet users')[['Country Name', 'Internet users']]
print("Top 5 Countries with Highest Internet Usage:\n", top_5_countries)
```

<h1 style='color:blue'>Insight 3</h1>
<h3>The chart shows that high-income countries are the most numerous, followed by lower-middle and upper-middle-income countries, while low-income countries are the least represented. This suggests that economic disparities exist, with fewer countries in the lowest income bracket compared to other categories.</h3>

```python
plt.figure(figsize=(8, 5))
sns.countplot(y=info['Income Group'], palette='viridis', order=info['Income Group'].value_counts().index)
plt.title('Count of Countries by Income Group')
plt.xlabel('Number of Countries')
plt.ylabel('Income Group')
plt.show()
```

<h1 style='color:blue'>Insight 4</h1>
<h3>The box plot indicates that the median internet usage among countries is around 50%, with a wide interquartile range suggesting significant variation in internet access. The whiskers extend close to 0% and 100%, implying that some countries have very low or nearly universal internet penetration.</h3>

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=info['Internet users'], color='orange')
plt.title('Box Plot of Internet Usage')
plt.xlabel('Internet Users (%)')
plt.show()
```

<h1 style='color:blue'>Insight 5</h1>
<h3>The dataset shows no countries with zero internet access, suggesting global improvements. As an alternative, we identified countries where internet usage is below 20%, highlighting regions that still face connectivity challenges and may need further support to bridge the digital divide.</h3>

```python
no_internet_countries = info[info['Internet users'] == 0][['Country Name']]
print("Countries with No Internet Access:\n", no_internet_countries)

#Here the country with low internet(below 20)
low_internet_countries = info[info['Internet users'] < 20][['Country Name', 'Internet users']]
print("Countries with Low Internet Access (Below 20%):\n", low_internet_countries)
```

<h1 style='color:blue'>Insight 6</h1>
<h3>The dataset shows an average birth rate of 21.47, indicating a moderately high fertility rate. This suggests that many countries still experience significant population growth, which may impact economic development, healthcare, and resource planning.</h3>

```python
average_birth_rate = info['Birth rate'].mean()
print(f"Average Birth Rate: {average_birth_rate:.2f}")
```

<h1 style='color:blue'>Insight 7</h1>
<h3>The scatter plot visualizes the strong negative correlation between internet usage and birth rate. As internet penetration increases, birth rates tend to decline. This trend may reflect the influence of factors like education, access to healthcare, and economic development, which often accompany higher digital connectivity.</h3>

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=info['Internet users'], y=info['Birth rate'], color='red')
plt.title('Scatter Plot: Birth Rate vs. Internet Usage')
plt.xlabel('Internet Users (%)')
plt.ylabel('Birth Rate')
plt.show()
```

<h1 style='color:blue'>Insight 8</h1>
<h3>Several countries, primarily in Africa, have birth rates exceeding 40, with Niger having the highest at 49.66. High birth rates often indicate young populations and rapid population growth, which can impact healthcare, education, and economic development. These countries may require strategic planning to support their growing populations.</h3>

```python
high_birth_rate_countries = info[info['Birth rate'] > 40][['Country Name', 'Birth rate']]
print("Countries with Birth Rates Above 40:\n", high_birth_rate_countries)
```

<h1 style='color:blue'>Insight 9</h1>
<h3>The pie chart illustrates the distribution of countries across different income groups. The varying proportions highlight global economic disparities, with some regions having a higher concentration of low-income countries while others dominate the high-income category. Understanding this distribution helps in assessing economic development and policy-making priorities.</h3>

```python
plt.figure(figsize=(8, 5))
info['Income Group'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'gold', 'red'])
plt.title('Proportion of Income Groups')
plt.ylabel('')
plt.show()
```

<h1 style='color:blue'>Insight 10</h1>
<h3>The strong negative correlation (-0.82) between birth rate and internet usage suggests that countries with higher birth rates tend to have lower internet penetration. This may indicate that factors such as economic development, education, and infrastructure influence both digital access and population growth trends.</h3>

```python
correlation = info['Birth rate'].corr(info['Internet users'])
print(f"Correlation between Birth Rate and Internet Usage: {correlation:.2f}")
```
