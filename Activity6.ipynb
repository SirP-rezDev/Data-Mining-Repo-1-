```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

import warnings
warnings.filterwarnings('ignore')

dengue = pd.read_csv('dengue.csv')
dengue
```

<h1 style='color: blue'>Insight 1</h1>
<h3>The month of <span style='color: maroon; font-wieght: bold;'>August</span> recorded the highest average number of dengue cases in a year, indicating a peak in dengue incidence during this period.</h3>

```python
monthly_avg_cases = dengue.groupby("Month")["Dengue_Cases"].mean()
month_order = [
    "January", "February", "March", "April", "May", "June", 
    "July", "August", "September", "October", "November", "December"
]
monthly_avg_cases = monthly_avg_cases.reindex(month_order)
highest_avg_month = monthly_avg_cases.idxmax()
highest_avg_cases = monthly_avg_cases.max()

print(f"The month with the highest average dengue cases is {highest_avg_month} with an average of {highest_avg_cases:.2f} cases.")

plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_avg_cases.index, y=monthly_avg_cases.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Average Dengue Cases")
plt.title("Average Dengue Cases per Month")
plt.axhline(highest_avg_cases, color="red", linestyle="--", label=f"Highest: {highest_avg_month}")
plt.legend()
plt.show()
```

<h1 style='color:blue'>Insight 2</h1>
<h3>The National Capital Region (NCR) recorded the highest number of dengue-related deaths, with a total of <span style='color: maroon'>4008</span> deaths. This highlights a critical need for enhanced dengue prevention and control measures in the area. </h3>

```python
region_death_cases = dengue.groupby("Region")["Dengue_Deaths"].sum()

highest_death_region = region_death_cases.idxmax()
highest_death_cases = region_death_cases.max()

print(f'The region with the highest death case is "{highest_death_region}" with "{highest_death_cases} deaths" in whole year')

plt.figure(figsize=(12, 6))
sns.barplot(x=region_death_cases.index, y=region_death_cases.values, palette="Reds_r")
plt.xticks(rotation=45)
plt.xlabel("Region")
plt.ylabel("Total Dengue Deaths")
plt.title("Total Dengue Deaths by Region")
plt.axhline(highest_death_cases, color="black", linestyle="--", label=f"Highest: {highest_death_region}")
plt.legend()
plt.show()
```

<h1 style='color:blue'>Insight 18</h1>
<h3>The barplot shows that dengue cases are highest in autumn and summer, while they are lowest in spring. This suggests that warmer and more humid conditions during these seasons may create a favorable environment for mosquito breeding, increasing the risk of dengue transmission.</h3>

```python
season_mapping = {
    "December": "Winter", "January": "Winter", "February": "Winter",
    "March": "Spring", "April": "Spring", "May": "Spring",
    "June": "Summer", "July": "Summer", "August": "Summer",
    "September": "Autumn", "October": "Autumn", "November": "Autumn"
}
dengue["Season"] = dengue["Month"].map(season_mapping)
seasonal_cases = dengue.groupby("Season")["Dengue_Cases"].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=seasonal_cases.index, y=seasonal_cases.values, palette="coolwarm")
plt.xlabel("Season")
plt.ylabel("Average Dengue Cases")
plt.title("Seasonal Trend of Dengue Cases")
plt.show()
```

<h1 style='color:blue'>Insight 21</h1>
